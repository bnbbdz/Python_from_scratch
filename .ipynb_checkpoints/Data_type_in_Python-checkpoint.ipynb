#### Indexing string / slicing / stride / concatenate

HELLO WARS

0123456789

Indexing
```sql
--SQL: 
select substr('Hello wars',1,1) , substr('Hello wars',6,1)  from dual
```

````python
#Python
indx = 'Hello wars'
print(indx[0],indx[6],indx[-1],indx[9],sep='\n')
````

Slicing
```sql
--SQL: 
select substr('Hello wars',1,4)  from dual
```

````python
#Python
print(indx[0:5],indx[:5],indx[6:],sep='\n')
````

Stride
```sql
--SQL: 
select substr('Hello wars',1,4,2)  from dual
```

````python
#Python
print(indx[0:5:2])
````

Concatenate Strings
```sql
--SQL: 
select 'Hello world ' || ' - ' || 'Hello wars'  from dual
```

````python
#Python
new_string_1 = var + ' - ' + indx
print('String 1: ', new_string_1)

new_string_2 = f'{var} - {indx}'
print('String 2: ' , new_string_2)

new_string_3 = '{0} - {1}'.format(var,indx)
print('String 3: ' , new_string_3)
````

Convert all the characters in string to upper case
```sql
--SQL: 
select upper('Hello wars')  from dual
```

````python
#Python
print(var.upper())
print(var.lower())
print(var.capitalize())
print(var.casefold())
print(var.title())
````

Replace the old substring with the new target substring is the segment has been found in the string
```sql
--SQL: 
select replace('SQL is the best','SQL','Python') from dual
```

````python
#Python
a = "SQL is the best"
b = a.replace('SQL', 'Python')
b
````

````python
#Only the index of the first elment of substring in string will be the output

print(a.find('is'))

#other way
print(a.index('is'))

print(a.find('vip')) #if not match give -1

print(a.index('vip')) #if not match -> error
````

````python
#Combination
print('Case 1: ',True and True)
print('Case 2: ',True or False)
print('Case 3: ',True and False)
print('Case 4: ',False or False)
print('Case 5: ',True and False and False or True)
print('Case 6: ',True and (False and (False or True)))
````

````python
#Try to convert other types to bool:
print('Case 1: ',bool(''))
print('Case 2: ',bool(' '))
print('Case 3: ',bool(None))
print('Case 4: ',bool(1))
print('Case 5: ',bool(0))
````

Check if a char is digit

````python
check_d = '123'
print('Is alphabet: ',check_d.isalpha())
print('Is alphabet & number: ',check_d.isalnum())
print('Is digit: ',check_d.isdigit())
print('Is numeric: ',check_d.isnumeric())
````

````python
check_d = 'abc'
print('Is alphabet: ',check_d.isalpha())
print('Is alphabet & number: ',check_d.isalnum())
print('Is digit: ',check_d.isdigit())
print('Is numeric: ',check_d.isnumeric())
````

````python
check_d = 'abc123'
print('Is alphabet: ',check_d.isalpha())
print('Is alphabet & number: ',check_d.isalnum())
print('Is digit: ',check_d.isdigit())
print('Is numeric: ',check_d.isnumeric())
````

### Datetime

```sql
--SQL: 
select sysdate from dual
```

````python
#Python
import datetime as dt
today = dt.datetime.now() #dt.datetime.today()
today
````

```sql
--SQL: 
select to_char(sysdate,'DD') from dual
```

````python
#Python
print('Datetime',today,'Day',today.day,'Month',today.month,'Year',today.year,
        'Hour',today.hour,'Minute',today.minute,'Time',today.time(),
        'Date',today.date(),sep='\n')
````

Add date

```sql
--SQL: 
select sysdate + 1 from dual
```

````python
#Python
print(today + dt.timedelta(days = 1))
````

Format datetime

```sql
--SQL: 
select to_char(sysdate,'YYYYMMDD') from dual
```

````python
#Python
print(dt.datetime.strftime(today,'%Y%m%d'))
````

Convert char to datetime

```sql
--SQL: 
select to_date('20220323','YYYYMMDD') from dual
```

````python
#Python
print(dt.datetime.strptime('20220323','%Y%m%d'))
````
