#### Left Join table

```sql
with tb1 as
(
select ORDERNUMBER,QUANTITYORDERED as QUANTITYORDERED_1
from dbo.data_sample
where ORDERNUMBER = '10411'
	and QUANTITYORDERED <= 30
),
tb2 as
(
select ORDERNUMBER,QUANTITYORDERED as QUANTITYORDERED_2
from dbo.data_sample
where ORDERNUMBER = '10411'
	and QUANTITYORDERED > 30
)
select a.ORDERNUMBER
		,QUANTITYORDERED_1
		,QUANTITYORDERED_2
		,QUANTITYORDERED_1 - QUANTITYORDERED_2 NET_QUANTITYORDERED
from tb1 a
left join tb2 b
on	a.ORDERNUMBER = b.ORDERNUMBER
```

Create temp table

````python
tb1 = data_sample[['ORDERNUMBER','QUANTITYORDERED']][(data_sample['ORDERNUMBER'] == 10411) & (data_sample['QUANTITYORDERED'] <= 30)]
tb2 = data_sample[['ORDERNUMBER','QUANTITYORDERED']][(data_sample['ORDERNUMBER'] == 10411) & (data_sample['QUANTITYORDERED'] > 30)]
````

````python
tb1
````

````python
tb2
````

````python
join_tb = pd.merge(tb1,tb2, on='ORDERNUMBER' , how='left')
````

````python
join_tb.head()
````

````python
join_tb.rename(columns={'QUANTITYORDERED_x':'QUANTITYORDERED_1','QUANTITYORDERED_y':'QUANTITYORDERED_2'},inplace=True)
````

````python
join_tb['NET_QUANTITYORDERED'] = join_tb['QUANTITYORDERED_1'] - join_tb['QUANTITYORDERED_2']
join_tb
````

Line number

```sql
select ROW_NUMBER() over(partition by ORDERNUMBER order by QUANTITYORDERED desc) so_thu_tu
	,a.ORDERNUMBER,a.QUANTITYORDERED,a.PRICEEACH
from dbo.data_sample a
where ORDERNUMBER ( '10100','10101')
```

````python
df_101 = data_sample[['ORDERNUMBER','QUANTITYORDERED','PRICEEACH']][(data_sample['ORDERNUMBER'] == 10100) | (data_sample['ORDERNUMBER'] == 10101)]
````

````python
df_101
````

````python
#ROW_NUMBER() over(partition by ORDERNUMBER order by QUANTITYORDERED desc) so_thu_tu
df_101['SO_THU_TU'] = df_101.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].rank(method = 'first', ascending = False)
df_101 = df_101.sort_values(by=['ORDERNUMBER','SO_THU_TU'])
````

````python
df_101
````

Running on total

```sql
select ROW_NUMBER() over(partition by ORDERNUMBER order by QUANTITYORDERED desc) so_thu_tu
	,a.ORDERNUMBER,a.QUANTITYORDERED,a.PRICEEACH
	,sum(a.QUANTITYORDERED) over(partition by ORDERNUMBER order by QUANTITYORDERED desc) running_on_total
from dbo.data_sample a
where ORDERNUMBER IN ( '10100','10101')
```

````python
#Running on total
df_101['RUNNING_ON_TOTAL'] = df_101.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].cumsum()
````

````python
df_101
````

```sql
select ROW_NUMBER() over(partition by ORDERNUMBER order by QUANTITYORDERED desc) so_thu_tu
	,a.ORDERNUMBER,a.QUANTITYORDERED,a.PRICEEACH
	,sum(a.QUANTITYORDERED) over(partition by ORDERNUMBER order by QUANTITYORDERED desc) running_on_total
	,avg(a.QUANTITYORDERED) over(partition by ORDERNUMBER) average_quantity
from dbo.data_sample a
where ORDERNUMBER IN ( '10100','10101')
```

````python
df_101['AVERAGE_QUANTITY'] = df_101.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].transform('mean')
````
